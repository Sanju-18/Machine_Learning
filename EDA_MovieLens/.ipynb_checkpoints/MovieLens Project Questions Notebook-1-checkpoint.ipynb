### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python
df_item['release_year'] = df_item['release date'].apply(lambda x: int(x.split('-')[2]))
plt.figure(figsize=(20,8))
temp = pd.DataFrame(df_item.groupby('release_year').sum().loc[:,'Action':].reset_index().set_index('release_year'))
temp = pd.melt(temp.reset_index(), id_vars=['release_year'],value_vars=temp.columns)
data = temp.pivot_table(index='variable', columns='release_year', values='value')
sns.heatmap(data)
```

### 9. Find the top 25 movies, each having a minimum of 100 ratings - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

#1.First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.

```python
df = pd.merge(df_data, df_item, how='left', on=['movie id'])
df_movies = df.groupby('movie title')['rating'].count().to_frame('rating').reset_index()
movies_list = df_movies[df_movies.rating>100]['movie title'].to_list()
```

#2.Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.

```python
df = pd.merge(df_data, df_item, how='left', on=['movie id'])
df = df.groupby('movie title')['rating'].mean().sort_values(ascending = False).to_frame('avg rating').reset_index()
```

#3.Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

```python
top_df = df[pd.DataFrame(df['movie title'].tolist()).isin(movies_list).any(1)].reset_index()
print('Top 25 movies, which has more than 100 reviews:',top_df.head(25)['movie title'].tolist())
```
